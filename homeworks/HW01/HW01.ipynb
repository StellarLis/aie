### Примеры создания массивов:

```python
arr = np.array([1, 2, 3, 4]) # Создание массива
print(arr)
```

```python
# Этот же массив можно получить используя np.arange()
arr = np.arange(1, 5)
print(arr)
```

```python
arr = np.ones(4) # Создание массива из единичек
arr
```

```python
arr = np.eye(4) # Создание единичной матрицы размером 4
arr
```

```python
np.empty((3, 3)) # Создание пустого массива (т.е. заполненного мусором из памяти)
```

### Примеры векторных операций и базовых статистик

```python
a = np.array([3, 4, 5, 6])
a
```

```python
# Создание массива из одного числа (такого не было на семинаре, я сам нашёл в интернете)
b = np.full(4, 2) # 4 - shape, 2 - число, которым заполниться массив
b
```

```python
a * b # Поэлементное умножение
```

```python
b * 2 # Умножение массива на скаляр
```

```python
b = np.array([10, 20, 30, 40])
a / b # Поэлементное деление
```

```python
a + b # Поэлементное сложение
```

```python
# Демонстрация средней величины (mean) и разных других штук
demo_array = np.array([10, 20, 30, 40, 50])
print('Средняя величина:', demo_array.mean())
print('Минимальный элемент:', demo_array.min())
print('Максимальный элемент:', demo_array.max())

# Нормы (По сути мы определяем каждое значение на отрезке от 0 до 1)
norm = (demo_array - demo_array.min()) / (demo_array.max() - demo_array.min())
print('Нормы:', norm)
```
